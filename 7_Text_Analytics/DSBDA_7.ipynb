#Text Analytics

1 Extract sample document and apply following document preprocessing techniques: Tokenization, POS Tagging, Stop Words Removal, Stemming, and Lemmatization.

2 Create representation of document by calculating Term Frequency.



```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from collections import Counter
```

```python
# Download required resources (run once per session)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Sample text
text = "Natural Language Processing is an important part of Data Science."
```

```python
# 1. Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
# 2. POS Tagging
pos = pos_tag(tokens)
print("POS Tags:", pos)
```

```python
# 3. Stopwords Removal
stop_words = set(stopwords.words('english'))
filtered = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("Filtered Tokens:", filtered)
```

```python
# 4. Stemming
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in filtered]
print("Stemmed Words:", stemmed)
```

```python
# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered]
print("Lemmatized Words:", lemmatized)
```

```python
# 6. Term Frequency
tf = Counter(lemmatized)
print("Term Frequency:", tf)
```
