# Data Visualization I

1 Use the inbuilt dataset ‘titanic’. The dataset contains 891 rows and information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data.

2 Write a code to check how the price of the ticket (column name: ‘fare’) for each passenger is distributed by plotting a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load inbuilt dataset
df = sns.load_dataset('titanic')
```

```python
# Plot histogram
sns.histplot(df['fare'], bins=30, kde=True)
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Count")

# Show plot
plt.show()
```

Observation : “The fare distribution is right-skewed. Most passengers paid lower fares, while a few paid higher fares.”
