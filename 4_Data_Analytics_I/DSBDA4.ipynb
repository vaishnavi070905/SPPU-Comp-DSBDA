# Data Analytics I
Create a Linear Regression Model using Python or R to predict the value of prices of houses using the Boston Housing Dataset  
(available at https://www.kaggle.com/c/boston-housing (kaggle.com in Bing)).

The dataset provides information about various houses in Boston through different attributes such as crime rate, number of rooms, accessibility to highways, etc., and contains 14 feature variables.

Objective:  
To predict the value of house prices using the given dataset.

```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('HousingData.csv')
```

```python
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# Handle missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
X
```

```python
Y = df['MEDV']
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
#Train the Model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
train_Score = model.score(X_train, Y_train)
print("Train Score : ", train_Score)
```

```python
Test_Score = model.score(X_test, Y_test)
print("Test Score : ", Test_Score)
```

```python
import numpy as np
#Evaluation RMSE

np.sqrt(mean_squared_error(Y_test, y_pred))
```
