#Data Visualization II

1 Use the inbuilt dataset ‘titanic’ as used in the above problem. Plot a box plot for distribution of age with respect to each gender, along with the information about whether they survived or not.
(Column names: ‘sex’ and ‘age’)

2 Write observations on the inference from the above statistics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = sns.load_dataset('titanic')
```

```python
# Plot boxplot
sns.boxplot(x='sex', y='age', hue='survived', data=df)

# Labels
plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

# Show plot
plt.show()
```

Observation : “The boxplot shows that female passengers had a higher survival rate compared to males. Younger passengers also had better chances of survival. Male passengers show wider age variation and lower survival.”
