```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title("Age Distributated by Gender and survival")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```
